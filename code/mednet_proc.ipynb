```python
b1, i1 = run('pat_data', ['classification/2'], [[
    'mednet_features'
    ]], print, plot_flag=True, feature_selection=True, cal_baselines=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
b = [np.max([b1['svc'][i], b1['xgboost'][i], b1['mlp'][i]]) for i in range(len(b1['svc']))]
df = pd.DataFrame({'Baseline': b, 'MLP': i1['mlp'], 'XgBoost': i1['xgboost'], 'SVC': i1['svc']})
df = pd.DataFrame(columns=['type', 'auc', 'seed'])
for i in range(len(b)):
    df = df.append({'type': 'Baseline', 'auc': b[i], 'seed': i}, ignore_index=True)
    df = df.append({'type': 'MLP', 'auc': i1['mlp'][i], 'seed': i}, ignore_index=True)
    df = df.append({'type': 'XgBoost', 'auc': i1['xgboost'][i], 'seed': i}, ignore_index=True)
    df = df.append({'type': 'SVC', 'auc': i1['svc'][i], 'seed': i}, ignore_index=True)
sns.barplot(data=df, x='seed', y='auc', hue='type')
plt.ylim(0.7, 1.0)
plt.ylabel('AUC')
plt.show()
```

```python
print('Best Baseline mean: {}, std: {}'.format(np.mean(b), np.std(b)))
print('SVM mean: {}, std: {}'.format(np.mean(i1['svc']), np.std(i1['svc'])))
print('XgBoost mean: {}, std: {}'.format(np.mean(i1['xgboost']), np.std(i1['xgboost'])))
print('MLP mean: {}, std: {}'.format(np.mean(i1['mlp']), np.std(i1['mlp'])))
```

```python
from scipy.stats import ttest_rel
print('SVM vs Baseline: {}'.format(ttest_rel(i1['svc'], b, alternative='greater').pvalue))
print('XgBoost vs Baseline: {}'.format(ttest_rel(i1['xgboost'], b, alternative='greater').pvalue))
print('MLP vs Baseline: {}'.format(ttest_rel(i1['mlp'], b, alternative='greater').pvalue))
```
